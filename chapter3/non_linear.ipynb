# Non Linear Transformations

The real world is filled with non-linearities and so dealing with it often requires non-linear computation. This demo shows how to compute non-linear functions using nengo. The two non-linear functions demonstrated in this demo are 'multiplication' and 'squaring'.

The model takes two inputs. For multiplication, we essentially project both inputs independently into a 2D space, and then decode a nonlinear transformation of that space (the product of the first and second vector elements). The model also squares the value of the first input (inputA) encoded in a population (A), in the output of a another population (C). 

```python
import matplotlib.pyplot as plt
%matplotlib inline

import nengo
```

## Step 1: Create the model

The model has four ensembles: two input ensembles ('A' and 'B'), a 2D combined ensemble ('Combined'), and an output ensemble ('D').

```python
# Create the model object
model = nengo.Network(label='Multiplication')
with model:
    # Create 4 ensembles of leaky integrate-and-fire neurons
    A = nengo.Ensemble(100, dimensions=1, radius=1)
    B = nengo.Ensemble(100, dimensions=1, radius=1)
    C = nengo.Ensemble(100, dimensions=1, radius=1)
    combined = nengo.Ensemble(224, dimensions=2, radius=2) # This radius is ~sqrt(10^2+10^2)
    prod = nengo.Ensemble(100, dimensions=1, radius=1)

# This next two lines make all of the encoders in the Combined population point at the 
# corners of the cube. This improves the quality of the computation.
# Note the number of neurons is assumed to be divisible by 4
from nengo.dists import Choice
# Comment out the line below for 'normal' encoders
combined.encoders = Choice([[1,1],[-1,1],[1,-1],[-1,-1]])
```

##Step 2: Create input for the model

We will use two varying scalar values for the two input signals that drive activity in ensembles A and B.

```python
from nengo.utils.functions import piecewise
with model:
    # Create piecewise step functions for input   
    """inputA = nengo.Node(piecewise({0: -0.75, 1.6: 0.5, 3.2: -0.75}))
    inputB = nengo.Node(piecewise({0: 1, 1.6: 0.25, 3.2: -0.25}))
    
    actual_mul = piecewise({0: -0.75, 1.6: 0.125, 3.2: 0.1875})
    actual_sq = piecewise({0: 0.5625, 1.6: 0.25, 3.2: 0.5625})"""
    
    inputA = nengo.Node(piecewise({0: -0.75, 1.25: 0.5, 2.5: -0.75, 3.75: 0}))
    inputB = nengo.Node(piecewise({0: 1, 1.25: 0.25, 2.5: -0.25, 3.75: 0.75}))
    
    actual_mul = piecewise({0: -0.75, 1.25: 0.125, 2.5: 0.1875, 3.75: 0})
    actual_sq = piecewise({0: 0.5625, 1.25: 0.25, 2.5: 0.5625, 3.75: 0})
```

## Step 3: Connect the elements of the model

```python
with model:
    # Connect the input nodes to the appropriate ensembles
    nengo.Connection(inputA, A)
    nengo.Connection(inputB, B)
    
    # Connect input ensembles A and B to the 2D combined ensemble
    nengo.Connection(A, combined[0])
    nengo.Connection(B, combined[1])
    
    # Define a function that computes the multiplication of two inputs
    def product(x):
        return x[0] * x[1]
    
    # Define the squaring function
    def square(x):
        return x[0] * x[0]
    
    # Connect the combined ensemble to the output ensemble D
    nengo.Connection(combined, prod, function=product)
    
    # Connection ensemble A to ensemble C
    nengo.Connection(A, C, function=square)
```

##Step 4: Add probes to collect data

Collect output data from each ensemble and input.

```python
with model:
    inputA_probe = nengo.Probe(inputA)
    inputB_probe = nengo.Probe(inputB)
    A_probe = nengo.Probe(A, synapse=0.01)
    B_probe = nengo.Probe(B, synapse=0.01)
    C_probe = nengo.Probe(C, synapse=0.1)
    combined_probe = nengo.Probe(combined, synapse=0.01)
    prod_probe = nengo.Probe(prod, synapse=0.01)
```

## Step 5: Run the model

```python
# Create the simulator
sim = nengo.Simulator(model)
# Run it for 5 seconds
sim.run(5)
```

## Step 6: Plot the results

To check the performance of the model, we can plot the input signals and decoded ensemble values. 

```python
# Plot the input signals and decoded ensemble values for multiplication
plt.plot(sim.trange(), sim.data[A_probe], label="Decoded Ensemble A")
plt.plot(sim.trange(), sim.data[B_probe], label="Decoded Ensemble B")
plt.plot(sim.trange(), sim.data[prod_probe], label="Decoded product")
plt.plot(sim.trange(), [actual_mul(t) for t in sim.trange()], c='k', label="Actual Product")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.ylim(-1, 1.5);

# Plot the input signal and decoded ensemble values for Squaring
plt.figure()
plt.plot(sim.trange(), sim.data[A_probe],  label='Decoded Ensemble A')
plt.plot(sim.trange(), sim.data[C_probe], 'r', label='Decoded Square')
plt.plot(sim.trange(), [actual_sq(t) for t in sim.trange()], c='k', label="Actual Square")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.ylim(-1, 1);
```

The input signals we chose makes it obvious when things are working. When input_A is zero, both the product and the square should be zero as well.
